```python
import numpy as np
import numpy.linalg as npl
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
import seaborn as sns
```

```python
sns.set_style('darkgrid')

# Define two vectors in R^2
a = np.array([2, 3])
b = np.array([1, 1])

# Vector addition (component-wise)
c = a + b
print(f'{a} + {b} =', c)

# Scalar multiplication
lambda_ = 2
scaled_a = lambda_ * a

# Plotting
plt.figure(figsize=(8, 6))

# Plotting vectors a and b
plt.quiver(0, 0, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='a')
plt.quiver(0, 0, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='g', label='b')

# Plotting vector c (a + b)
plt.quiver(0, 0, c[0], c[1], angles='xy', scale_units='xy', scale=1, color='b', label='a + b')

# Plotting scaled vector of a
plt.quiver(0, 0, scaled_a[0], scaled_a[1], angles='xy', scale_units='xy', scale=1, color='y', label='lambda * a')

# Setting the limits and labels
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.title('Vector Operations in R^2')
plt.show()
```

Ex. 2.2 (Systems of linear equations)

```python
matrix = np.vstack([[1,1,1],[1,-1,2],[2,0,3]])
solution = np.vstack([3,2,1])

#attempt to solve the system of equations
try:
    npl.solve(matrix,solution)
except:
    print('No solution🥱')
    print(np.hstack([matrix,solution]))
```

```python
matrix = np.vstack([[1,1,1],[1,-1,2],[0,1,1]])
solution = np.vstack([3,2,2])

#attempt to solve the system of equations
try:
    solve = npl.solve(matrix,solution)
    #if successful, print the solution
    print('Solution found')
    print(np.hstack([matrix,solution]))
    print(solve)
except:
    print('No solution🥱')
    print(np.hstack([matrix,solution]))
```
